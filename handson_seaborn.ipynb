```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Tips dataset
tips = sns.load_dataset("tips")
```

"Tips" dataset records information about restaurant bills and tips, along with a
attributes like day of the week, time, and other factors.

```python
tips.head()
```

Univariate Analysis: Univariate analysis is a statistical procedure that examines a single variable in a dataset to describe and summarize it. It's used to explore the range of values, central tendency, and response pattern for each variable.


Let's first look at individual variables. We'll examine the distribution of total_bill and tip to understand how often certain values appear in the dataset.

```python
plt.figure(figsize=(8, 6))
sns.histplot(tips['total_bill'], kde=True)
plt.title("Distribution of Total Bill")
plt.show()
```

 From this plot, we might observe that most bills fall in a certain range (e.g., \$10\$30).



```python
plt.figure(figsize=(8, 6))
sns.histplot(tips['tip'], kde=True, color='orange')
plt.title("Distribution of Tip Amount")
plt.show()
```

 Bivariate Analysis:
Next, we analyze how two variables interact. We can look at the relationship between total_bill and tip to see if higher bills tend to result in higher tips.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Total Bill vs. Tip")
plt.show()
```

This plot reveals a positive relationship, where larger bills often correspond to larger tips.

Categorical Analysis

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='sex', y='tip', data=tips)
plt.title("Tip Amount by gender")
plt.show()
```

Multivariate Analysis

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='day', y='tip', hue='sex', data=tips, split=True)
plt.title("Tip Amount by Day and Gender")
plt.show()
```

**Insights & Conclusion: **
Using Seaborn for data visualization allowed us to extract several insights
from the Tips dataset:

* There is a positive relationship between total bill amount and tip amount.
* Tipping behavior may vary by gender and day of the week.
